```python
import ollama
import time


modelos = ['phi3', 'qwen2.5:1.5b']

prompt_prueba = """Eres un asistente estricto. Redacta UN SOLO mensaje de Telegram.
REGLAS ESTRICTAS:
1. MÁXIMO 3 LÍNEAS.
2. Tono amigable y persuasivo.
3. Ofrece un 15% de descuento por renovación anticipada.
4. NUNCA menciones riesgos, abandono, ni probabilidades.

Contexto del cliente:
- Nombre: Carlos
- Problema: Queda 1 mes de contrato y sus visitas cayeron un 60%.

Mensaje de Telegram:"""


for modelo in modelos:
    print(f"Modelo: {modelo.upper()}")
    try:
        inicio = time.time()
        
        respuesta = ollama.chat(model=modelo, messages=[
            {'role': 'user', 'content': prompt_prueba}
        ], options={'temperature': 0.1}) 
        
        tiempo_total = time.time() - inicio
        mensaje = respuesta['message']['content']
        
        print(f"Tiempo: {tiempo_total:.2f} segundos")
        print(f"{mensaje}\n{'-'*40}")
        
    except Exception as e:
        print(f"Error: {e}")
```
